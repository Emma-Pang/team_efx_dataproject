```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import random
```

```python
# Files to Load 
overdose_df = pd.read_csv("csv_series/overdose-death-rates-per-population.csv", sep = ";")

overdose_df.head()
```

```python
unique_st = overdose_df['Location'].unique()
unique_st.sort()
print(unique_st)
```

```python
# #group data by states
# grouped_overdose_bystate = overdose_df.groupby(["Location"])

state_dfs_dictionary = {}
state_ids = set(overdose_df['Location'])

for state in state_ids:
    state_dfs_dictionary[state] = overdose_df[overdose_df['Location'] == state]
    
state_dfs_dictionary['District of Columbia']
```

```python
for st in unique_st:
    filename = "csv_series/overdose_"+str(st)+".csv"
    state_dfs_dictionary[st].to_csv(filename, index=False, header=True)
```
